````python
from graphviz import Digraph

def trace(root):
    # builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    # graohviz to create the graph
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right

    # trace enumerates through all the nodes and edges 
    nodes, edges = trace(root)
    for node in nodes:
        print("grad while tracing: ", node.grad, "node label: ", node.label)
    # print("nodes: ", nodes)
    # print("edges: ", edges)
    for n in nodes:
        uid = str(id(n))
        print("grad while drawing: ", n.grad)
        dot.node(name=uid, label="{ %s | data: %.4f | grad: %.4f }" % (n.label, n.data, n.grad), shape='record')
        if n._op:
            # if this value is a result of some operation, create an op node for it
            dot.node(name = uid + n._op, label = n._op)
            # and connect this node to it
            dot.edge(uid + n._op, uid)

    for n1, n2 in edges:
        # connect n1 to the op node of n2
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)

    return dot
````

````python
class Value:
    # we add tanh operation
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0 # initial value - zero means there is no effect of the value of the node on the output intially
        self._prev = set(_children)
        self._op = _op
        self.label = label
        
    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')
        return out
````

````python
class Value():

    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0
        self._backward = lambda: None # in default it is a function that doesn't do anything
        self._prev = set(_children)
        self._op = _op
        self.label = label

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            # based on chain rule we multiply the gradient of the output dy/dz = dy/dx * dx/dz
            self.grad = out.grad * 1.0
            other.grad = out.grad * 1.0
        out._backward = _backward
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')

        def _backward():
            self.grad = out.grad * 1.0
            other.grad = out.grad * 1.0
        out._backward = _backward
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad = out.grad * other.data
            other.grad = out.grad * self.data
        out._backward = _backward
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')

        def _backward():
            self.grad = out.grad * (1 - out.data**2)
            print("grad in tanh: ", self.grad)
            # other.grad = out.grad * self.data
        out._backward = _backward
        return out
````

## Automatic Backpropogation

We don't want to calculate the backward on each node, instead once all the forward calculations for a node are done, we want to call the _backward() on it. This ordering of the graph can be acheived through topological sort - a laying out of the graph so that all the edges only go from left to right

https://venkys.io/articles/details/topological-sort

Topological sort is a useful algorithm in graph theory that orders the vertices of a directed graph in such a way that for every directed edge (u, v), vertex u comes before vertex v in the ordering. It is often used to solve problems that have dependencies between tasks or events.

The algorithm starts by identifying vertices with no incoming edges and adds them to the sorted list. Then, it removes these vertices and their outgoing edges from the graph. This process is repeated until all vertices have been processed.

Topological sort has applications in various fields, such as project management, task scheduling, and resolving dependencies in software development. It helps identify the order in which tasks or events should be executed to satisfy all dependencies.

## Topological Sort Code
Basically keep going to child of the current node till there are no more children left, visited is a set because we don't want any multiple visits to be recorded and once wwe get to absoluted leaf nodes that have no children then we add them and recursively adding all the nodes after that

````python
topo = []
visited = set()
def build_topo(v):
    # starts at some root node, o in this case
    if v not in visited:
        visited.add(v)
        for child in v._prev: # we go through all of its children and we lay them out from left to right
            build_topo(child)
        topo.append(v) # we only the root node itself, once all the children are processed
build_topo(o)
print(topo)
````

## Solution to this error

https://en.wikipedia.org/wiki/Chain_rule#Multivariable_case

basically we accumulate these gradients - i.e., these gradients add up

````python
class Value():

    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0
        self._backward = lambda: None # in default it is a function that doesn't do anything
        self._prev = set(_children)
        self._op = _op
        self.label = label

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            # chain rule fix for multivariable case 
            # gradients accumulate
            self.grad += out.grad * 1.0
            other.grad += out.grad * 1.0
        out._backward = _backward
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')

        def _backward():
            self.grad = out.grad * 1.0
            other.grad = out.grad * 1.0
        out._backward = _backward
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad += out.grad * other.data
            other.grad += out.grad * self.data
        out._backward = _backward
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')

        def _backward():
            self.grad = out.grad * (1 - out.data**2)
            print("grad in tanh: ", self.grad)
            # other.grad = out.grad * self.data
        out._backward = _backward
        return out

    def backward(self):
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        build_topo(self)
        self.grad = 1.0
        for node in reversed(topo):
            node._backward()
````

````python
class Value():

    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0
        self._backward = lambda: None # in default it is a function that doesn't do anything
        self._prev = set(_children)
        self._op = _op
        self.label = label

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # fix: make other an instance of Value
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            self.grad += out.grad * 1.0
            other.grad += out.grad * 1.0
        out._backward = _backward
        return out

    def __sub__(self, other):
        # fix: make other an instance of Value
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data - other.data, (self, other), '-')

        def _backward():
            self.grad = out.grad * 1.0
            other.grad = out.grad * 1.0
        out._backward = _backward
        return out

    def __mul__(self, other):
        # fix: make other an instance of Value
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad += out.grad * other.data
            other.grad += out.grad * self.data
        out._backward = _backward
        return out

    def __rmul__(self, other):
        # acts like a fallback 
        # if python can't do 2 * a it will check by any chance can it to the reverse and do that instead
        return self * other

    def __div__(self, other):
        # fix: make other an instance of Value
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def exp(self):
        x = self.data
        out = Value(np.exp(x), (self, ), 'exp')
        def _backward():
            self.grad = out.grad * out.data # out.data already stores the value of $e^x$
        out._backward = _backward

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')

        def _backward():
            self.grad = out.grad * (1 - out.data**2)
            print("grad in tanh: ", self.grad)
            # other.grad = out.grad * self.data
        out._backward = _backward
        return out

    def backward(self):
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        build_topo(self)
        self.grad = 1.0
        for node in reversed(topo):
            node._backward()
````

## division fix

Division can be decomposed as follows:
```bash
a / b
a * (1/b)
a * (b**-1)
```

we want create the mother of all functions to handle this case of $b^{-1}$ which is $x^k$

````python
class Value():

    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0
        self._backward = lambda: None # in default it is a function that doesn't do anything
        self._prev = set(_children)
        self._op = _op
        self.label = label

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # fix: make other an instance of Value
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            self.grad += out.grad * 1.0
            other.grad += out.grad * 1.0
        out._backward = _backward
        return out

    def __neg__(self):
        return self * -1
        
    def __sub__(self, other):
        # we use negation and addition (stuff we've already built to implement this function)
        return self + (-other)

    def __mul__(self, other):
        # fix: make other an instance of Value
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad += out.grad * other.data
            other.grad += out.grad * self.data
        out._backward = _backward
        return out
        
    def __pow__(self, other):
        # fix: basically the differential for anything other than a float or int will be different
        assert isinstance(other, (int, float)), "only supporting int or float for now"
        out = Value(self.data**other, (self, ), f"**{other}")
        
        def _backward():
            self.grad += out.grad * other * (self.data ** (other - 1))
        out._backward = _backward
        return out

    def __rmul__(self, other):
        # acts like a fallback 
        # if python can't do 2 * a it will check by any chance can it to the reverse and do that instead
        return self * other

    def __truediv__(self, other):
        return self * other**-1

    def exp(self):
        x = self.data
        out = Value(math.exp(x), (self, ), 'exp')
        def _backward():
            self.grad += out.grad * out.data # out.data already stores the value of $e^x$
        out._backward = _backward
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')

        def _backward():
            self.grad = out.grad * (1 - out.data**2)
            print("grad in tanh: ", self.grad)
            # other.grad = out.grad * self.data
        out._backward = _backward
        return out

    def backward(self):
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        build_topo(self)
        self.grad = 1.0
        for node in reversed(topo):
            node._backward()
````

````python
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')
b = Value(6.8813735870195432, label='b')
x1w1 = x1*w1; x1w1.label = 'x1w1'
x2w2 = x2*w2; x2w2.label = 'x2w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1w1x2w2'
n = x1w1x2w2 + b; n.label = 'n'

o = n.tanh(); o.label = 'o'
o.backward()
draw_dot(o)
````
