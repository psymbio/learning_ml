## Creating a Graph-like Structure

Now all we are missing is the connective tissue between these values - we want to create a graph to understand which values contribute to which other values and how we can propogate through the graph to understand the chain reaction of change that takes place.

````python
class Value:

    # we introduce a value children which is a set
    # it is a set for efficiency
    def __init__(self, data, _children=()):
        self.data = data
        self._prev = set(_children)
        
    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the children
        out = Value(self.data + other.data, (self, other))
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other))
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other))
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other))
        return out

a = Value(2.0)
b = Value(-3.0)
c = Value(10.0)
d = a * b + c
# -6.0 + 10.0
print(d._prev)
````

````python
a = Value(2.0)
b = Value(-3.0)
c = Value(10.0)
d = (a * b) + (c * a) + b
# -6.0 + 20.0 - 3.0
# 14.0 - 3.0
print(d._prev)
````

But we also need to know the operation that is being done

````python
class Value:

    # we add operation
    def __init__(self, data, _children=(), _op=''):
        self.data = data
        self._prev = set(_children)
        self._op = _op
        
    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

a = Value(2.0)
b = Value(-3.0)
c = Value(10.0)
d = a * b + c
# -6.0 + 10.0
print(d._prev, d._op)
````

## Visualizing the Expressions

Because these expressions shall become much larger than this - we want to have a way to visualize them.

````python
from graphviz import Digraph

def trace(root):
    # builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right

    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        dot.node(name = uid, label = "{ data %.4f }" % (n.data), shape='record')
        if n._op:
            # if this value is a result of some operation, create an op node for it
            dot.node(name = uid + n._op, label = n._op)
            # and connect this node to it
            dot.edge(uid + n._op, uid)

    for n1, n2 in edges:
        # connect n1 to the op node of n2
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)

    return dot
````

````python
draw_dot(d)
````

3. Graph Construction
The computational graph is constructed using Graphviz.

Function trace(root)
Traverses the computation graph starting from root (d in this case), storing:

+ Nodes: All Value objects encountered.
+ Edges: Parent-child relationships between values.
+ Function draw_dot(root)
+ Generates a Graphviz DAG (Directed Acyclic Graph):

Each Value is represented as a node with data. If a Value was created via an operation, a separate node represents the operation. Arrows (edge) connect dependencies.

Graph Representation for d = a * b + c
```markdown

          |
        ( + )
          |
        (4.0)
```
Each operation (* and +) has its own node, and arrows represent the computation flow.

## Backpropogation

Now, we want to visualize the backpropagation - we want to calculate the derivative of each of these nodes w.r.t $L$


## Basic Derivative Concept

- The derivative of $L$ with respect to itself $\frac{dL}{dL}$ is **1**.
- This serves as a fundamental identity in differentiation.

## Computing Derivatives with Respect to Other Variables

- We can compute the derivative of $L$ with respect to other variables, such as:
  - $f$
  - $d$
  - $c$
  - $e$
  - $a$
  - $b$

These derivatives help us understand how changes in these variables impact the function $L$.

## Application in Neural Networks

- In neural networks, we compute derivatives to determine how weights and biases influence the overall function.
- This process is essential for **backpropagation**, where we adjust weights and biases to minimize loss and improve model performance.

## Implementing a Gradient Value in Each Node

We create a grad in each node which maintains the derivative of the value w.r.t $L$.

````python
class Value:

    # we add operation
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0 # initial value - zero means there is no effect of the value of the node on the output intially
        self._prev = set(_children)
        self._op = _op
        self.label = label
        
    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')
e = a * b; e.label = 'e'
d = e + c; d.label = 'd'

f = Value(-2.0, label='f')
L = d * f; L.label = 'L'

from graphviz import Digraph

def trace(root):
    # builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    # graohviz to create the graph
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right

    # trace enumerates through all the nodes and edges 
    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        dot.node(name=uid, label="{ %s | data: %.4f | grad: %.4f }" % (n.label, n.data, n.grad), shape='record')
        if n._op:
            # if this value is a result of some operation, create an op node for it
            dot.node(name = uid + n._op, label = n._op)
            # and connect this node to it
            dot.edge(uid + n._op, uid)

    for n1, n2 in edges:
        # connect n1 to the op node of n2
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)

    return dot

draw_dot(d)
````

## Understanding Derivatives

We start with the given equation:

$$L = d \cdot f$$

To find the derivative of $L$ with respect to $d$, we use the definition of derivatives:

$$\frac{dy}{dx} = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$$

here $x$ is the term against which we want to differentiate against i.e., we want to see the slope of change if this particualr term is changes by a small amount.

Substituting $L = d \cdot f$:

$$\frac{dL}{dd} = \lim_{h \to 0} \frac{(d + h) f - d f}{h}$$

Expanding:

$$\frac{dL}{dd} = \lim_{h \to 0} \frac{d f + h f - d f}{h} = \lim_{h \to 0} \frac{h f}{h}$$

Since $h$ cancels out:

$$\frac{dL}{dd} = f$$

Thus, the derivative of $L$ with respect to $d$ is simply $f$.

Symmetrically,

$$\frac{dL}{df} = d$$

## Given Expressions

We define the relationships between variables as follows:

$$e = a \cdot b$$

$$d = e + c$$

$$L = d \cdot f$$

## Finding $\frac{dL}{de}$

Using the chain rule, we differentiate $L$ with respect to $e$:

$$\frac{dL}{de} = \frac{d}{de} \big( f \cdot d \big)$$

Since $d = e + c$, we substitute this into the equation:

$$L = f \cdot d = f \cdot (e + c)$$

Now, applying the definition of the derivative:

$$\frac{dL}{de} = \lim_{h \to 0} \frac{L(e + h) - L(e)}{h}$$

Substituting $L = f \cdot (e + c)$:

$$\frac{dL}{de} = \lim_{h \to 0} \frac{f \cdot (e + h + c) - f \cdot (e + c)}{h}$$

Expanding:

$$\frac{dL}{de} = \lim_{h \to 0} \frac{f e + f h + f c - (f e + f c)}{h}$$

Canceling out terms:

$$\frac{dL}{de} = \lim_{h \to 0} \frac{f h}{h}$$

Since $h$ cancels out:

$$\frac{dL}{de} = f$$

## Another Way of Finding $\frac{dL}{de}$

Using the chain rule, we differentiate $L$ with respect to $e$:

$$\frac{dL}{de} = \frac{d}{de} \big( d \cdot f \big)$$

Applying the product rule:

$$\frac{dL}{de} = f \cdot \frac{d}{de} (d)$$

Since $d = e + c$, its derivative with respect to $e$ is:

$$\frac{d}{de} (e + c) = 1$$

Thus, we obtain:

$$\frac{dL}{de} = f \cdot 1 = f$$

## Conclusion

The derivative of $L$ with respect to $e$ is **$f$**.

https://en.wikipedia.org/wiki/Chain_rule

## dL/dc normal derivation and then another derivation with a complete explanation of the chain rule

## 1. Direct Differentiation Approach

We start with the given equation:

$$L = f \cdot d$$

Since we know that:

$$d = e + c$$

Substituting this into $L$:

$$L = f \cdot (e + c)$$

Now, differentiating $L$ with respect to $c$:

$$\frac{dL}{dc} = \frac{d}{dc} \big( f \cdot (e + c) \big)$$

Since $f$ and $e$ are independent of $c$, we apply the derivative:

$$\frac{dL}{dc} = f \cdot \frac{d}{dc} (e + c)$$

The derivative of $e + c$ with respect to $c$ is simply 1:


$$\frac{dL}{dc} = f \cdot 1 = f$$

Thus, the derivative of $L$ with respect to $c$ is:

$$\frac{dL}{dc} = f$$

---

## 2. Derivation Using the Chain Rule

The chain rule states that if a function $L$ depends on an intermediate variable $d$, which in turn depends on $c$, then the derivative of $L$ with respect to $c$ is:

$$\frac{dL}{dc} = \frac{dL}{dd} \cdot \frac{dd}{dc}$$

### Step 1: Compute $\frac{dL}{dd}$

From the given equation:

$$L = f \cdot d$$

Differentiating both sides with respect to $d$:

$$\frac{dL}{dd} = f$$

### Step 2: Compute $\frac{dd}{dc}$

Since:

$$d = e + c$$

Differentiating both sides with respect to $c$:

$$\frac{dd}{dc} = 1$$

### Step 3: Apply the Chain Rule

Now using the chain rule:

$$\frac{dL}{dc} = \frac{dL}{dd} \cdot \frac{dd}{dc}$$

Substituting the values:

$$\frac{dL}{dc} = f \cdot 1 = f$$

### Conclusion

Thus, using both direct differentiation and the chain rule, we find:

$$\frac{dL}{dc} = f$$

This confirms that both approaches lead to the same result.


## Understanding the Chain Rule as a Multiplication of Rates of Change

### **Step 1: Concept of Rates of Change**
A derivative measures how one quantity changes in response to another. For example:
- $\frac{dL}{dd}$ tells us how $L$ changes with respect to $d$.
- $\frac{dd}{dc}$ tells us how $d$ changes with respect to $c$.

If $L$ depends on $d$, and $d$ depends on $c$, then the total change of $L$ with respect to $c$ can be found by **multiplying these intermediate rates of change**:

$$
\frac{dL}{dc} = \frac{dL}{dd} \cdot \frac{dd}{dc}
$$

As put by George F. Simmons: "If a car travels twice as fast as a bicycle and the bicycle is four times as fast as a walking man, then the car travels 2 × 4 = 8 times as fast as the man."

---

### **Step 2: Intuitive Explanation**
Think of this in real-world terms:

1. Suppose you're driving a car and want to measure how your **position ($L$) changes** over time, but you only have information about:
   - **Speed ($\frac{dL}{dd}$)** – how your position changes per unit of time.
   - **Time per fuel unit ($\frac{dd}{dc}$)** – how much time passes per unit of fuel consumed.

2. If you want to determine how your position changes **with respect to fuel consumption ($c$)**, you multiply:

   $$
   \frac{\text{distance}}{\text{time}} \times \frac{\text{time}}{\text{fuel}} = \frac{\text{distance}}{\text{fuel}}
   $$

   This shows that the effect of one small change in fuel ($c$) on distance traveled ($L$) is determined by multiplying the rate at which fuel affects time and the rate at which time affects distance.

---

### **Step 3: Mathematical Derivation**
Let’s formally derive the chain rule:

1. Assume $L$ is a function of $d$, and $d$ is a function of $c$, i.e.:

   $$
   L = f(d), \quad d = g(c)
   $$

2. The derivative of $L$ with respect to $c$ is:

   $$
   \frac{dL}{dc} = \lim_{h \to 0} \frac{f(g(c+h)) - f(g(c))}{h}
   $$

3. To express this in terms of $d$, introduce an intermediate step:

   $$
   \frac{dL}{dc} = \lim_{h \to 0} \left( \frac{f(g(c+h)) - f(g(c))}{g(c+h) - g(c)} \times \frac{g(c+h) - g(c)}{h} \right)
   $$

4. As $h \to 0$, the first term approaches $\frac{dL}{dd}$ and the second term approaches $\frac{dd}{dc}$:

   $$
   \frac{dL}{dc} = \frac{dL}{dd} \times \frac{dd}{dc}
   $$

This proves the **chain rule** mathematically and explains why it works as a multiplication of intermediate rates of change.

---

### **Conclusion**
The chain rule works because **the total rate of change of one quantity with respect to another is the product of all intermediate rates of change**. Just like in real-world scenarios where speed and fuel consumption rates combine to determine total distance per fuel unit, in calculus, derivatives "chain together" by multiplication. 🚀


## Understanding Backpropagation Using the Chain Rule

Backpropagation is a method used in neural networks to efficiently compute gradients. It relies on the **chain rule** to propagate errors from the output layer backward through the network. The key idea is that **to compute the gradient at any node, you only need to know the gradient of the next node and the local derivative at the current node**.

---

### **1. Backpropagation as a Flow of Gradients**
In a neural network, we have multiple layers where each node (neuron) depends on values from the previous layer. When computing gradients for weight updates, the process flows **backward**, starting from the loss at the output layer.

Using the chain rule, the gradient of the loss ($L$) with respect to an earlier node ($x$) is found by multiplying:
- The gradient of the loss with respect to the next node ($y$), i.e., $\frac{dL}{dy}$.
- The local derivative $\frac{dy}{dx}$, which tells how $y$ changes with respect to $x$.

Thus, the total gradient at $x$ is:

$$
\frac{dL}{dx} = \frac{dL}{dy} \cdot \frac{dy}{dx}
$$

This means **if we know the gradient at a later node, we can compute the gradient at the current node just by multiplying by the local derivative**.

---

### **2. Backpropagation in a Computational Graph**
Consider a simple neural network with a sequence of operations:

1. **Forward Pass (Computing Output)**
   - Each neuron receives inputs, applies a transformation (e.g., a weighted sum followed by an activation function), and passes the result to the next layer.

2. **Backward Pass (Computing Gradients)**
   - We start from the final loss $L$ and move backward, applying the chain rule at each step.

Let's take an example:

#### **Example: Single Path in a Computational Graph**
Suppose we have:

$$
z = w \cdot x + b
$$

$$
y = f(z)
$$

$$
L = \text{loss}(y, \hat{y})
$$

To update $w$, we need $\frac{dL}{dw}$. Using the chain rule:

$$
\frac{dL}{dw} = \frac{dL}{dy} \cdot \frac{dy}{dz} \cdot \frac{dz}{dw}
$$

Each term in this multiplication corresponds to:
1. **$\frac{dL}{dy}$** → The gradient of the loss w.r.t. the output of the neuron.
2. **$\frac{dy}{dz}$** → The derivative of the activation function.
3. **$\frac{dz}{dw} = x$** → The local derivative of $z$ w.r.t. $w$.

Since each gradient depends only on the next layer, **we can compute gradients sequentially from the output to the input**.

---

### **3. Gradient Flow in a Multi-Layer Network**
For a deep network, backpropagation works by:
1. **Computing the loss gradient at the output layer**.
2. **Passing the gradients backward layer by layer**, multiplying by local derivatives at each step.
3. **Each neuron only needs to know the gradient of the next node and the local derivative** to compute its own gradient.

For example, in a network with multiple hidden layers:

$$
h_1 = f_1(x)
$$

$$
h_2 = f_2(h_1)
$$

$$
h_3 = f_3(h_2)
$$

$$
L = \text{loss}(h_3, \hat{y})
$$

The gradient of $L$ w.r.t. the first layer input $x$ is:

$$
\frac{dL}{dx} = \frac{dL}{dh_3} \cdot \frac{dh_3}{dh_2} \cdot \frac{dh_2}{dh_1} \cdot \frac{dh_1}{dx}
$$

Each term is computed by moving backward through the network, applying the chain rule at each step.

---

### **4. Why Backpropagation is Efficient**
- We **reuse** gradients from the next layer instead of computing from scratch.
- Each node **only requires local information**: the gradient from the next node and its local derivative.
- Computation is **layer-wise**, making it efficient for deep networks.

---

### **5. Summary**
- **Backpropagation is based on the chain rule**, which allows us to propagate gradients layer by layer.
- **Each node only needs to know the gradient from the next node and its own local derivative**.
- **The process flows backward**: first computing the loss gradient at the output and then moving step-by-step to the input layer.
- **Multiplication of intermediate derivatives ensures that each weight update is correctly computed**.

This is how neural networks learn efficiently using gradient descent.

## **1. Multilayer Perceptron (MLP)**

A **Multilayer Perceptron (MLP)** is a type of **feedforward neural network** consisting of multiple layers of neurons. It has at least **one hidden layer** and follows these key principles:

- **Input Layer**: Receives raw data.
- **Hidden Layers**: Process data using weights and activation functions.
- **Output Layer**: Produces the final prediction.

MLPs are **fully connected** networks, meaning every neuron in one layer connects to every neuron in the next layer.

### **MLP Architecture**
- A network with **one hidden layer** is called a **shallow network**.
- A network with **multiple hidden layers** is called a **deep neural network (DNN)**.

---

## **2. Neuron and Activation Function**
A **neuron** is the fundamental unit of an MLP. It performs the following operations:

1. Takes multiple inputs: $x_1, x_2, ..., x_n$.
2. Applies weights: $w_1, w_2, ..., w_n$.
3. Computes a weighted sum: 

   $$
   z = w_1x_1 + w_2x_2 + ... + w_nx_n + b
   $$

   where $b$ is the **bias term**.

4. Passes $z$ through an **activation function** to introduce non-linearity:

   $$
   a = f(z)
   $$

### **Common Activation Functions**
- **Sigmoid**: 

  $$
  f(z) = \frac{1}{1 + e^{-z}}
  $$

- **ReLU (Rectified Linear Unit)**:

  $$
  f(z) = \max(0, z)
  $$

- **Tanh**:

  $$
  f(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}
  $$

- **Softmax (for classification tasks)**:

  $$
  f(z_i) = \frac{e^{z_i}}{\sum_{j} e^{z_j}}
  $$

Each activation function helps the network **capture complex patterns** in data.

---

### **Key Takeaways**
- **MLPs are fully connected neural networks with multiple layers.**
- **Each neuron computes a weighted sum of inputs and applies an activation function.**
- **Activation functions introduce non-linearity, allowing the network to learn complex relationships.**
- **The backpropagation algorithm is used to train MLPs by adjusting weights.**

MLPs are widely used for classification and regression tasks in machine learning!

````python
class Value:
    # we add tanh operation
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0 # initial value - zero means there is no effect of the value of the node on the output intially
        self._prev = set(_children)
        self._op = _op
        self.label = label
        
    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')
        return out
````

## Derivative of $e^{2x}$ Using First Principles

Using the first principle of derivatives, we differentiate $e^{2x}$:

$$\frac{d}{dx} e^{2x} = \lim_{h \to 0} \frac{e^{2(x + h)} - e^{2x}}{(x + h) - x}$$

### Step 1: Expanding the Exponential Term
Using the exponent property $e^{a + b} = e^a \cdot e^b$, we rewrite:

$$\frac{d}{dx} e^{2x} = \lim_{h \to 0} \frac{e^{2x} \cdot e^{2h} - e^{2x}}{h}$$

### Step 2: Factoring Out $e^{2x}$
$$\frac{d}{dx} e^{2x} = \lim_{h \to 0} \frac{e^{2x} \cdot (e^{2h} - 1)}{h}$$

### Step 3: Splitting the Limit
$$\frac{d}{dx} e^{2x} = e^{2x} \cdot \lim_{h \to 0} \frac{e^{2h} - 1}{h}$$

### Step 4: Evaluating the Limit
Using the known limit formula:

$$\lim_{h \to 0} \frac{e^h - 1}{h} = 1$$

We substitute $h$ with $2h$:

$$\lim_{h \to 0} \frac{e^{2h} - 1}{h} = 2 \lim_{h \to 0} \frac{e^{2h} - 1}{2h} = 2 \cdot 1 = 2$$

### Step 5: Final Result
$$\frac{d}{dx} e^{2x} = e^{2x} \cdot 2$$

$$\frac{d}{dx} e^{2x} = 2e^{2x}$$

## Derivation of the Known Limit Formula:
$$\lim_{h \to 0} \frac{e^h - 1}{h} = 1$$

### Step 1: Consider the Definition of $e^h$
Using the Maclaurin series expansion for $e^h$:

$$e^h = 1 + h + \frac{h^2}{2!} + \frac{h^3}{3!} + \frac{h^4}{4!} + \dots$$

Subtracting 1 from both sides:

$$e^h - 1 = h + \frac{h^2}{2!} + \frac{h^3}{3!} + \frac{h^4}{4!} + \dots$$

### Step 2: Divide by $h$
$$\frac{e^h - 1}{h} = \frac{h + \frac{h^2}{2!} + \frac{h^3}{3!} + \dots}{h}$$

Splitting the terms:

$$\frac{e^h - 1}{h} = 1 + \frac{h}{2!} + \frac{h^2}{3!} + \frac{h^3}{4!} + \dots$$

### Step 3: Apply the Limit as $h \to 0$
As $h$ approaches 0, all terms involving $h$ vanish:

$$\lim_{h \to 0} \frac{e^h - 1}{h} = 1 + 0 + 0 + 0 + \dots = 1$$

### Conclusion:
Thus, we have proved:

$$\lim_{h \to 0} \frac{e^h - 1}{h} = 1$$


## Derivative of $e^{2x}$ Using the Chain Rule

An alternative and easier way to derive the derivative of $e^{2x}$ is by using the **chain rule**.

### Step 1: Define the Function
Let:
$$f(x) = e^{2x}$$
This is an exponential function where the exponent itself is a function of $x$.

### Step 2: Use the Chain Rule
The chain rule states:
$$\frac{d}{dx} e^{g(x)} = e^{g(x)} \cdot g'(x)$$
where $g(x) = 2x$ in our case.

### Step 3: Differentiate the Inner Function
$$g(x) = 2x \quad \Rightarrow \quad g'(x) = 2$$

### Step 4: Apply the Chain Rule
$$\frac{d}{dx} e^{2x} = e^{2x} \cdot 2$$

### Step 5: Final Answer
$$\frac{d}{dx} e^{2x} = 2e^{2x}$$

This approach is much simpler than using first principles.

## Derivative of $e^{2x}$ Using the Product Rule

We use the property of exponentials:

$$e^{a + b} = e^a \cdot e^b$$

Let’s rewrite $e^{2x}$ as a product:

$$e^{2x} = e^x \cdot e^x$$

### Step 1: Define $u$ and $v$
Let:
- $u = e^x$, so $u' = e^x$
- $v = e^x$, so $v' = e^x$

### Step 2: Apply the Product Rule
The product rule states:

$$\frac{d}{dx} [u \cdot v] = u'v + uv'$$

Substituting $u$ and $v$:

$$\frac{d}{dx} e^{2x} = e^x \cdot e^x + e^x \cdot e^x$$

$$\frac{d}{dx} e^{2x} = 2e^{2x}$$

Thus, the derivative of $e^{2x}$ using the product rule is:

$$\frac{d}{dx} e^{2x} = 2e^{2x}$$


## Derivative of $\tanh(x)$ Using the Quotient Rule

We start with the exponential form of $\tanh(x)$:

$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

### Step 1: Define $u$ and $v$
Let:
- $u = e^x - e^{-x}$, so $u' = e^x + e^{-x}$
- $v = e^x + e^{-x}$, so $v' = e^x - e^{-x}$

### Step 2: Apply the Quotient Rule
The quotient rule states:

$$\frac{d}{dx} \left( \frac{u}{v} \right) = \frac{u' v - u v'}{v^2}$$

Substituting $u, u', v, v'$:

$$\frac{d}{dx} \tanh(x) = \frac{(e^x + e^{-x})(e^x + e^{-x}) - (e^x - e^{-x})(e^x - e^{-x})}{(e^x + e^{-x})^2}$$

### Step 3: Expand the Numerator
Using the identity $(a+b)^2 - (a-b)^2 = 4ab$:

$$(e^x + e^{-x})^2 - (e^x - e^{-x})^2 = 4e^x e^{-x} = 4$$

So the numerator simplifies to:

$$4$$

### Step 4: Simplify the Expression
Since the denominator is $(e^x + e^{-x})^2$, we write:

$$\frac{d}{dx} \tanh(x) = \frac{4}{(e^x + e^{-x})^2}$$

### Final Answer
$$\frac{d}{dx} \tanh(x) = 1 - \tanh^2(x)$$

or in terms of $e^x$:

$$\frac{d}{dx} \tanh(x) = \frac{4}{(e^x + e^{-x})^2}$$

## Derivative of \( \tanh(x) \) Using the Quotient Rule

We start with the given expression:

\[
\tanh(x) = \frac{e^{2x} - 1}{e^{2x} + 1}
\]

### Step 1: Define \( u \) and \( v \)
Let:
- \( u = e^{2x} - 1 \), so \( u' = 2e^{2x} \)
- \( v = e^{2x} + 1 \), so \( v' = 2e^{2x} \)

### Step 2: Apply the Quotient Rule
The quotient rule states:

\[
\frac{d}{dx} \left( \frac{u}{v} \right) = \frac{u' v - u v'}{v^2}
\]

Substituting \( u, u', v, v' \):

\[
\frac{d}{dx} \tanh(x) = \frac{(2e^{2x})(e^{2x} + 1) - (e^{2x} - 1)(2e^{2x})}{(e^{2x} + 1)^2}
\]

### Step 3: Expand the Numerator

\[
2e^{2x} (e^{2x} + 1) - 2e^{2x} (e^{2x} - 1)
\]

Expanding both terms:

\[
(2e^{4x} + 2e^{2x}) - (2e^{4x} - 2e^{2x})
\]

\[
2e^{4x} + 2e^{2x} - 2e^{4x} + 2e^{2x}
\]

\[
4e^{2x}
\]

### Step 4: Simplify the Expression

Since the denominator remains:

\[
(e^{2x} + 1)^2
\]

We get:

\[
\frac{d}{dx} \tanh(x) = \frac{4e^{2x}}{(e^{2x} + 1)^2}
\]

### Step 5: Express in Terms of \( \tanh(x) \)

Using the identity:

\[
\sech^2(x) = \frac{4e^{2x}}{(e^{2x} + 1)^2}
\]

We conclude:

\[
\frac{d}{dx} \tanh(x) = 1 - \tanh^2(x)
\]

This is the standard derivative of \( \tanh(x) \).
