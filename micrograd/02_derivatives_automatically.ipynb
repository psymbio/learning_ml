```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from graphviz import Digraph

def trace(root):
    # builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    # graohviz to create the graph
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right

    # trace enumerates through all the nodes and edges 
    nodes, edges = trace(root)
    print("nodes: ", nodes)
    print("edges: ", edges)
    for n in nodes:
        uid = str(id(n))
        dot.node(name=uid, label="{ %s | data: %.4f | grad: %.4f }" % (n.label, n.data, n.grad), shape='record')
        if n._op:
            # if this value is a result of some operation, create an op node for it
            dot.node(name = uid + n._op, label = n._op)
            # and connect this node to it
            dot.edge(uid + n._op, uid)

    for n1, n2 in edges:
        # connect n1 to the op node of n2
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)

    return dot
```

```python
class Value:
    # we add tanh operation
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0 # initial value - zero means there is no effect of the value of the node on the output intially
        self._prev = set(_children)
        self._op = _op
        self.label = label
        
    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')
        return out
```

```python
# inputs x1, x2

# inputs
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')

# weights
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')

# bias of the neuron
b = Value(6.8813735870195432, label='b')

# doing x1*w1 + x2*w2 + b step-wise
x1w1 = x1*w1; x1w1.label = 'x1w1'
x2w2 = x2*w2; x2w2.label = 'x2w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1w1x2w2'
n = x1w1x2w2 + b; n.label = 'n'

o = n.tanh()

# local piece of chain rule
o.grad = 1.0
n.grad = 0.5
x1w1x2w2.grad = 0.5
x2w2.grad = 0.5
x1w1.grad = 0.5
w2.grad = x2w2.grad * x1.data
w1.grad = x2w2.grad * x2.data
x2.grad = x1w1.grad * w2.data
x1.grad = x1w1.grad * w1.data


# here, we get to see that wiggling the amount of w1 - shall have no effect on the final output 
# because the value it is multiplied with is zero
draw_dot(o)
```

```python
class Value():

    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0
        self.backward = lambda: None # in default it is a function that doesn't do anything
        self._prev = set(_children)
        self._op = _op
        self.label = label

    def __repr__(self):
        return f"Value(data={self.data}"

    def __add__(self, other):
        # while doing any calculation we feed in the operation that is being done
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            print(f"current node: {self.data}, out.grad: {out.grad}")
            self.grad = 1.0 * out.grad
            other.grad = 1.0 * out.grad
        out._backward = _backward
        return out

    def __sub__(self, other):
        out = Value(self.data - other.data, (self, other), '-')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad = out.grad * other.data
            other.grad = out.grad * self.data
        out._backward = _backward
        return out

    def __div__(self, other):
        out = Value(self.data / other.data, (self, other), '/')
        return out

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')

        def _backward():
            self.grad = out.grad * (1 - out.data**2)
            # other.grad = out.grad * self.data
        out._backward = _backward
        return out
```

```python
# inputs
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')

# weights
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')

# bias of the neuron
b = Value(6.8813735870195432, label='b')

# doing x1*w1 + x2*w2 + b step-wise
x1w1 = x1*w1; x1w1.label = 'x1w1'
x2w2 = x2*w2; x2w2.label = 'x2w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1w1x2w2'
n = x1w1x2w2 + b; n.label = 'n'

o = n.tanh()

o.grad = 1.0 # we need to be careful before carrying out a backward - o.grad is set to zero, we need to set it to 1
# differential for any value against itself is 1
o._backward()
n._backward()
b._backward()
x1w1x2w2._backward()
x2w2._backward()
x1w1._backward()
w2._backward()
w1._backward()
x2._backward()
x1._backward()

draw_dot(o)
```
