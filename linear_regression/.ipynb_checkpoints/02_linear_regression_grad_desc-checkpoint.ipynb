```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_function(a, x, b):
    return a*x + b
```

```python
a = 2
x = np.arange(100)
b = 45
y = linear_function(a, x, b)
```

```python
plt.scatter(x, y, s=10)
plt.show()
```

```python
# https://medium.com/@ms_somanna/guide-to-adding-noise-to-your-data-using-python-and-numpy-c8be815df524

noise = np.random.normal(0, 10, len(x))
y_noisy = y + noise
```

```python
plt.scatter(x, y_noisy, s=10)
plt.show()
```

```python
beta_0, beta_1 = 0, 1
alpha = 0.0001 # learning rate
iterations = 100
error = []

for i in range(iterations):
    y_pred = beta_0 * x + beta_1
    error = np.sum(np.power(y_pred - y_noisy, 2))
    print(error)
```
