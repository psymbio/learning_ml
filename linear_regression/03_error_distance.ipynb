```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data generation
np.random.seed(42)
x = np.linspace(0, 20, 25)
y_true = 2 * x + 1
y = y_true + np.random.normal(scale=3, size=x.shape)
mse = np.mean((y - y_true) ** 2)

# DataFrame for Seaborn
data = pd.DataFrame({'x': x, 'y': y, 'y_true': y_true})

# Plotting
plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")

# Scatter plot with error bars
sns.scatterplot(data=data, x='x', y='y', color='black', s=50, label='Measurements')
plt.plot(x, y_true, color='red', label='Underlying Truth', linewidth=2)

# Adding error lines
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y_true[i]], color='gray', linestyle='--', alpha=0.7)

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Linear Regression with Error Distances (MSE = {mse:.2f})')
plt.legend()
plt.savefig('error.png', transparent=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data generation
np.random.seed(42)
x = np.linspace(0, 20, 25)
y_true = 3 * x - 3  # Incorrect slope and intercept
y = 2 * x + 1 + np.random.normal(scale=3, size=x.shape)
mse = np.mean((y - y_true) ** 2)

# DataFrame for Seaborn
data = pd.DataFrame({'x': x, 'y': y, 'y_true': y_true})

# Plotting
plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")

# Scatter plot with error bars
sns.scatterplot(data=data, x='x', y='y', color='black', s=50, label='Measurements')
plt.plot(x, y_true, color='red', label='Incorrect Model', linewidth=2)

# Adding error lines
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y_true[i]], color='gray', linestyle='--', alpha=0.7)

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Incorrect Linear Regression with Error Distances (MSE = {mse:.2f})')
plt.legend()
plt.savefig('wrong_model.png', transparent=True)
plt.show()
```
